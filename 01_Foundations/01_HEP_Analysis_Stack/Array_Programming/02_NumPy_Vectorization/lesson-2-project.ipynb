In this file,

  * the `offsets` dataset has a cumulative sum of `counts`

### Charge quality cut

In the next part of the exercise, you'll apply quality cuts.

**First quality cut:** The two electrons are $e^+$ and $e^-$ and the two muons are $\mu^+$ and $\mu^-$.

<br>

Get the particle charges as arrays:

````python
e1_charge = np.asarray(dataset_hdf5["ee_mumu"]["e1"]["charge"])
e2_charge = np.asarray(dataset_hdf5["ee_mumu"]["e2"]["charge"])
mu1_charge = np.asarray(dataset_hdf5["ee_mumu"]["mu1"]["charge"])
mu2_charge = np.asarray(dataset_hdf5["ee_mumu"]["mu2"]["charge"])
````

<br>

Let's look at the distribution of charges in a 5 bin×5 bin histogram.

````python
from hist import Hist as hist

charge_hist = hist.new.Integer(-2, 3, name="q_ee").Integer(-2, 3, name="q_mumu").Double().fill(
    q_ee=e1_charge + e2_charge,
    q_mumu=mu1_charge + mu2_charge,
)
charge_hist
````

<br>

Most of the distribution is at `e1_charge + e2_charge == 0` and `mu1_charge + mu2_charge == 0` (high quality Higgs events).

We can ask the histogram for its bin data as an array:

````python
charge_hist.values()
````

Why is it distributed like that?

This concentration occurs because of charge conservation in the decay process. The Higgs boson is an electrically neutral particle, and it decays into two Z bosons, which are also neutral. When a Z boson decays into a pair of leptons (like an electron and a positron, or two muons), conservation of charge requires the pair to have a net charge of zero (+1+−1=0). Therefore, 'high-quality' signal events are expected to cluster at (0,0) in this histogram. Any events appearing at other values are likely 'background noise' random combinations of particles or misidentified tracks that do not originate from a Higgs decay.

<br>

We can also use Matplotlib to make overlays of cross-sections that do not include the central peak.

````python
import matplotlib.pyplot as plt
````

````python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

charge_hist[:, 0].plot(ax=ax1);
charge_hist[:, 4].plot(ax=ax1);
charge_hist[0, :].plot(ax=ax2);
charge_hist[4, :].plot(ax=ax2);
````

<br>

To select events with `e1_charge + e2_charge == 0` and `mu1_charge + mu2_charge == 0`, we can make each expression into a boolean array,

````python
e1_charge + e2_charge == 0
````

and apply it as a slice.

````python
fix, ax = plt.subplots(1, 1)

hist.new.Regular(120, 0, 120, name="mass Z_ee").Double().fill(
    zmass_ee
).plot(ax=ax);

hist.new.Regular(120, 0, 120, name="mass Z_ee").Double().fill(
    zmass_ee[e1_charge + e2_charge == 0]
).plot(ax=ax);
````

<br>

But how do you apply both?

Unfortunately, you have to use `&` for "and", `|` for "or", `~` for "not".

This is unfortunate because the comparison operation (`==`, `!=`, `<`, `>`, `<=`, `>=`) has to be surrounded by parentheses for the right order of operations.

````python
fix, ax = plt.subplots(1, 1)

hist.new.Regular(120, 0, 120, name="mass Z_mumu").Double().fill(
    zmass_mumu
).plot(ax=ax);

hist.new.Regular(120, 0, 120, name="mass Z_mumu").Double().fill(
    zmass_mumu[e1_charge + e2_charge == 0]
).plot(ax=ax);

hist.new.Regular(120, 0, 120, name="mass Z_mumu").Double().fill(
    zmass_mumu[(e1_charge + e2_charge == 0) & (mu1_charge + mu2_charge == 0)]
).plot(ax=ax);
````

## Exercise part 5

Make a final Higgs plot with and without the Z mass quality cut. It should look like this (100 bins):



**ANSWER HERE:**

````python
pass_charge = (e1_charge + e2_charge == 0) & (mu1_charge + mu2_charge == 0)
pass_z_cuts = (zmass_small > 12) & (zmass_small < 120) & (zmass_big > 40) & (zmass_big < 120)

selection = pass_charge & pass_z_cuts

h_all = Hist.new.Reg(100, 80, 160, name="m", label="$m_{4\ell}$ [GeV]").Double()
h_clean = Hist.new.Reg(100, 80, 160, name="m").Double()

h_all.fill(m=higgs_mass)
h_clean.fill(m=higgs_mass[selection])

fig, ax = plt.subplots(figsize=(10, 6))
h_all.plot(ax=ax, label="Datos Brutos (Sin cortes)", color="gray", alpha=0.3)
h_clean.plot(ax=ax, label="Señal Limpia (Cortes 2012)", color="blue", lw=2)

ax.legend()
plt.title("Búsqueda del Bosón de Higgs: $H \\rightarrow ZZ \\rightarrow 4\ell$")
plt.xlabel("$m_{4\ell}$ [GeV]")
plt.show()
````

<br><br><br><br><br>
